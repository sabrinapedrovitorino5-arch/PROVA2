```python

consumo_diario = [ 50, 45, 55, 60, 48, 52, 47, 49, 53, 51, 50, 46, 53, 61, 49, 53, 48, 52, 47, 56, 50, 55, 60, 48, 52, 47, 49, 53, 51 ]

lista = consumo_diario
```

```python
def consumo(lista):
    """
    Calcula a média de consumo diário de energia a partir de uma lista de valores.

    Args:
        lista (list): Uma lista de números inteiros ou floats representando
                      o consumo diário de energia em kWh.

    Returns:
        float: O consumo diário médio arredondado para duas casas decimais.
    """
    if not lista:
        return 0  

    soma_total = sum(lista)
    
    
    numero_dias = len(lista)
    
  
    media = soma_total / numero_dias
    
   
    return round(media, 2)
```

```python
media_consumo = consumo(lista)

print(f"O consumo diário médio foi: {media_consumo} kWh")
```
